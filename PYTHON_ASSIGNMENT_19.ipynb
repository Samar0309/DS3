## NOTE : IN OS.PATH>JOIN - YOU JUST GET THE PATH IN A STRING FORMAT , YOU ARE NOT CREATING ACTUAL FOLDERS

- YOU ARE JUST CONCATENATING A PATH AND OUTPUT WILL BE A NEW PATH IN STRING FORMAT

```python
curr_path = os.getcwd()
# here you need to save a current working directory/path in a variable for easy understanding
# now I also need a folder which I need to add or concatennate to my current path, so we will take that folder's name in a varible
folder1 = 'FOLDER 1'
join1 = os.path.join(curr_path,folder1) # this step will create a new path which we are saving in a variable called join1
print(join1) # DO NOT GO TO THE BELOW OUTPUT PATH and CHECK FOR FOLDER 1 , ITS IS NOT CREATED YET its JUST A STRING 
```

```python
# Now we will do for the FOLDER 3 and FOLDER 3

folder2 = 'FOLDER 2'
folder3 = 'FOLDER 3'

#        1.Now you can use original path to join FOLDER 1 + FOLDER 2 + FOLDER 3
#        OR
#        2. you can use already joined FOLDER 1 PATH

#     1  METHOD

joinfinal = os.path.join(curr_path,folder1,folder2,folder3) 
# this step will create a new path which we are saving in a variable called joinfinal
print(joinfinal)
```

## EASY MA
